```python
%matplotlib inline
import pandas as pd

ef=pd.read_csv("egd.csv",sep=",",nrows=10000)
```

```python
ef
```

```python
import networkx as nx
G=nx.from_pandas_edgelist(ef , 'id1','id2')
print(nx.info(G))
nx.draw(G,node_color='red')
```

```python
def initial_state(G):
    state = {}
    for node in G.nodes:
        state[node] = 'asleep'
    return state
```

```python
initial_state(G)
```

```python
import random

P_AWAKEN = 0.2
def state_transition(G, current_state):
    next_state = {}
    for node in G.nodes:
        if current_state[node] == 'asleep':
            if random.random() < P_AWAKEN:
                next_state[node] = 'awake'
    return next_state
```

```python
test_state = initial_state(G)
state_transition(G, test_state)
```

```python
from simulation import Simulation

sim = Simulation(G, initial_state, state_transition, name='Simple Sim')
```
